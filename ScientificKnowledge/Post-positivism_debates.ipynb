# Post-positivism Debates

Positivism is Knowledge that only comes from observable, measurable facts and scientific methods is meaningful (science = truth).  
Post-positivism or constructivism believes in science, but accepts that knowledge is never absolute and can always be revised (science = best available explaination so far).



 

#### Why exist?
1. Human cant know it perfectly (Economist can predict a market crash, but not timing).
2. Observations and data can be biased (Only ask urban people, ignoring rural opinions).
3. Science doesnot give final answer (time is the same, new are time is relative change by speed & gravity).
4. Science improve knowledge overtime (Research keep improving how we invest).
  
Post-positivism fixes science weaknesses, not rejects it.

#### Core Idea
A real world does exist, but we cant never know it perfectly & science give us approximate truth, not absolute truth.
1. Knowledge = Probabilistic
2. Conclusions = Tentative
3. Theories = Always revisable
```
There is no complete and perfect truth, but truth can be estimated.
```


#### Comparison
| Positivism | Post-positivism |
|------------|-----------------|
| One single reality/truth | Multiple interpretations of reality/truth |
| Objective certainty | Fallible, approximate knowledge |
| Facts are neutral (no good or bad) | Observations are theory laden (Influnced by what we already believe, expect, know) | 
| One best method (Science) | Multiple methods needed (Science, context, judgement, surveys) |

#### Critical realism 
Reality exists independently of us, but we can never know it perfectly. Science improves understanding by testing, correcting & retesting. All measurements may contain error & all theories may be wrong. Thats why post positivism :
1. Multiple methods
2. Multiple data sources
3. Multiple perspectives

#### Triangulation
Because no single method is perfect so post-positivism uses triangulation. Triangulation is using multiple ways to study the same phenomenon. Four types : 
1. Method triangulation : Quantitative + Qualitative
2. Data sources triangulation : Different people, times & settings
3. Analysis triangulation : Different analysts or model
4. Theory triangulation : Different theoretical lenses

#### Theory-Laden observation
Post-positivism argue that observations are never truly neutral (without any influenced ). What we see, notice, or measure is influenced by our background, culture, experience and beliefs.  

Post-positivism is not like relativism (Everything is subjective), There is reality/truth & people can reach the same understanding but it never 100% perfect. Bias cant be eliminated, but can be reduced using multiple methods (qualitative, quantitative), transparent assumption & peer review.
```
Observations are influenced by perspective, but reality/truth and  consensus are still possible.
```


#### Post-positivism debates in Social Science


1. Sources of knowledge : Observation + Theories + Context.
