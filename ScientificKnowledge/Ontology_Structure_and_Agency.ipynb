# Ontology

The study of nature of reality and being, what is real and what is exist in the field (physical, social, mental).
```
The belief about what kind of world you are studying (physical, social, mental).
```
It asking is reality one fixed thing, different for different people or always changing.

 #### Ontology in Social Science
In social science, we donâ€™t study rocks or atoms.
We study things like Culture, Power, Gender, Money, Education & Society. So the question becomes:
```
    Are these things real on their own, or are they created by people?
```

#### Ontological Views
1. Positivism : Only knowledge that comes from observable, measurable facts and scientific methods is meaningful (Quantitative)
2. Constructivism : Knowledge is subjective, socially constructed, and open to revision rather than absolute (Qualitative).
3. Pragmatism : Reality is not fixed, what matters is solvimg problems, use what works (Quantitative + Qualitative).

#### Ontology: Structure vs Agency
##### 1. Structure 
The system control you (law, economy, religion, education system, social class). These limit or guide what people do or can do.
```
Society is bigger than individuals, people are shaped by structures.
```
eg: School rules shape how students behave & Poor people have fewer opportunities because of the system
  ####
##### 2. Agency
People can control the system (Make decision by their ownself, free choice, individual decisions, human creativity). 
```
Society is created by people, humans are active decision-makers
```
eg: A person changes their life despite poverty & Social change starts from individual actions
  ####
##### 3. Modern View (Balance)
Most researchers now believe structures influence people, But people can also change structures.
```
Humans think, interpret, and give meaning to their actions
```
eg: Education system shapes students, But students can reform the education system.
