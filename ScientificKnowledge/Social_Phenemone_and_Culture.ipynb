# Social Phenemone and Culture


Any pattern of behavior, belief, interaction, or event that happens within a society and affects groups of people, not just individuals (Social media addiction, fashion trends, gender roles, economic inequality, online cancel culture & religious practices).
```
A social phenomenon is something people do, believe, or experience together, shaped by society.
```
  
```
Social phenomena are collective patterns of behavior shaped by culture, while culture is the shared system of meanings, values, and practices that guide how societies function.
```


##### Social
relating to how people interact, live together, and organize their collective life in groups, communities, and societies (Talking, family, religion & arguing).  

##### Society
people living together in a structured, lasting way, sharing rules, territory, and culture. [People, social interaction, rules & institutions].  

##### Culture
the way of life shared by a group of people (Beliefs, values, norms, language & traditions).


### Culture Shape Social Phenomena
Culture and social phenomena influence each other continuously. They are not separate, they exist in a feedback loop. Culture acts like a blueprint for how people behave in society : 
1. Values (what is important)
2. Norms (what is acceptable)
3. Beliefs (what is true or right)
4. Symbols (language, images)

#### Example :
##### Respect for Authority 
1. Cultural belief : Elders or leaders should not be questioned.
2. Behavior : Obedience & Limited criticism.
3. Social phenomenon : Hierarchical social structures.

## Criteria
### Criteria Social Phenomena
1. Collective : Involves groups, not just individuals.
2. Patterned : Happens repeatedly, not randomly.
3. Socially Influenced : Shaped by norms, rules, or expectations.
4. Observable : Can be seen, recorded, or studied.
5. Meaningful : Has shared meaning in society.
```
TikTok trends (velocity) → collective, repeated, socially shaped → social phenomenon
```  

### Criteria of Culture
1. Shared : Held by a group or society.
2. Learned : Passed through socialization, not genetics.
3. Symbolic : Uses language, signs, symbols.
4. Enduring : Lasts over time.
5. Dynamic : Can change and evolve
```
Language, traditions, customs, values
```

## Content
```
Content = information that shapes collective decision-making
```
### Content of Culture
1. Material Culture : Physical objects created by a society (Clothing, Technology, Architecture & Tools)
2. Non-Material Culture : Ideas and meanings (Beliefs, Narrative, Values, Norms, Language & Traditions)

### Content of Social Phenomena
1. Social behaviors (voting, protesting)
2. Social relationships (family, friendships)
3. Social institutions (education, religion)
4. Social roles (teacher, leader)
5. Social problems (poverty, discrimination)


### Operations
Operations = How they function, spread, and influence society
#### Operations Social Phenomena
1. Emergence : Arise from collective behavior.
2. Diffusion : Spread through communication and interaction.
3. Institutionalization : Become formal systems (laws, policies).
4. Transformation : Change culture or social structure.

#### Operations Culture
1. Socialization : People learn culture through family, school, media.
2. Norm Enforcement : Rewards for conformity & Punishment for deviance.
3. Transmission : Passed from generation to generation.
4. Adaptation : Changes in response to new conditions.

