# Science
Systematic methods to building knowledge through observation, experimentation, reasoning to make reliable explainations for the phenomena.  

Pure science = systematic investigation of nature, driven by curiousity & the desire to expand knowledge rather than by the intention to achieve practical outcomes.
```
Curiousity -> Investigation -> Expand
```
Applied science = Commonly defined as application & discoveries from pure science to address practical human needs (pure science leads to applied science).

##### Goals
1) Pure : Expand knowledge & find the truth.
2) Applied : Use knowledge with practical for a reason & needs.

### Core Components
#### 1. Observations
What you notice patterns in the world. Example prices move in trends, Volatility clusters & News affects markets.  

#### 2. Hypothesis
You propose a structured explanation. Example "Momentum predicts short-term returns" or "Negative sentiment leads to price drops” and A hypothesis must be testable.  

#### 3. Experiment / Test
You test the hypothesis under controlled conditions. Example backtesting a trading strategy, training a model on historical data & cross-validation.  

#### 4. Measurement & Error
Science cares more about how wrong you are than how right. Key ideas:
1. Noise
2. Bias
3. Variance
4. Overfitting
5. Uncertainty

#### 5. Revision
If the model fails:
1. Modify it
2. Replace it
3. Kill it

Failure is progress in science.


# Knowledge 
Understanding or awareness we gain about something  through learning, experience and thinking (Highest level of reality and truth, make the sense good or bad and right or wrong).

#### 1) Sources of Knowledge
1. Sense (sight, hearing, taste, touch, smell)
2. Positivist epistemology (Scientific based)
3. Other Sources (Insight, intuition, introspection, emphaty)

#### 2) Rational Intellect Knowledge
Thinking and reasoning part of human mind that allow to go beyond what we see, feel, to understand the meaning, cause, truth behind things (Question reality, analyze information, connect ideas, make logical conclusions).

#### 3) Type of Knowledge 
1) Religious knowledge : Origin of knowledge located beyond human control in omniscient force (God).
2) Science knowledge : Origin of knowledge are located within human being.

#### 4) Scientific knowledge 
Scientific knowledge is reliable knowledge produced through observation, experimentation, and reasoning that is continually tested and revised using empirical evidence.  

##### Type of Scientific Knowledge
1. Descriptive : Describes what happens (Prices are more volatile during news events).
2. Explanatory : Explains why it happens (Information shocks increase uncertainty).
3. Predictive : Predicts future outcomes (High volatility today → higher probability of volatility tomorrow). Prediction is the strongest test.
4. Applied : Used to change or control outcomes (Risk management, AI-driven trading & Engineering systems).

# Philosophy 
From greek philosophia "Love of Wisdom" is the systematically study of the most fundamental aspects of existence, knowledge, values, reason, language (intellectual discipline & way of life.
1. Philosophy have in every discipline (politics, religion, economy etc).
2. Every science is a child of philosophy - Alex Rosenberg 2012.

#### Major Brances of Philosophy
1. Meta physics : Study of reality, existence and nature of being (ontology & cosmology)
2. Epistemology : The study of knowledge, belief & justification (Rationalism, empiricism & constructivism)
3. Ethics : The study of morality, values & the good life (normative, meta & applied ethics)
4. Logic : The study of valid reasoning & argument (Classical logic, symbolic logic)

#### Philosophy of Science 
The study of how science works, what scientific knowledge is, and what makes science different from other ways of knowing.
