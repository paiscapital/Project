# Empiricism
Knowledge comes primarily from experience and observation of the world.
```
We know things because we observe them, not because they "feel true" & not because authority says so
```
##### Empiricism asks:
```
What does the data say?
```
  
##### Example 
We observe temperature, pressure, wind and we build models to predict rain. Even though we can't see future rain, empiricism allows theoretical models based on observed data.

### Core elements
1. Observation first : Knowledge begins with Seeing, Measuring & Recording.
2. Experience over speculation : Empiricism distrusts pure imagination & untestable theories & hypothesis.
3. Inductive reasoning : Empiricism uses induction Observe many cases & Infer general patterns (observation -> scientific method -> theory/conclusion).
4. Knowledge is probabilistic : Empirical knowledge is never 100% certain & always subject to revision.

### Tabula Rasa
John Locke introduced the idea of Tabula Rasa, which means "blank slate"
He believed that humans are born knowing nothing. At birth, the mind is empty, and all knowledge comes from experience and the senses.  
  
In contrast, Plato's essentialism argues that humans are born with built-in knowledge. Ideas such as right and wrong are already inside us, coded by nature or God, and learning is simply remembering what is already there.  
  
Aristotle believed that universal knowledge comes from observing particular experiences, not from inborn ideas.

# Positivism
Only knowledge that comes from observable, measurable facts and scientific methods is meaningful. Positivism is stricter than empiricism, only valid if it has measurable fact (weather).


### Core elements
1. Rejection of metaphysics : Positivism rejects spiritual explanations, moral claims as scientific & abstract speculation.
2. Science as the highest form of knowledge : Positivists believe Science is superior to philosophy, religion, or intuition for explaining reality.
3. Emphasis on measurement and laws : Positivism seeks regularities, laws & predictable patterns.
4. Objectivity : Knowledge should be neutral & free from values and emotions.

#### Strengths of Positivism
1. Strong rigor
2. Clear standards
3. Encourages precise measurement

#### Limits of Positivism
1. Ignores meaning, values, culture
2. Weak in social sciences
3. Can oversimplify complex human behavior
