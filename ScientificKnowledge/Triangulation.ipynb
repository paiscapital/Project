# Triangulation
Using multiple perspectives (methods, data, theories, or investigators) to study the same phenomenon in order to enhance credibility, validity, and depth of understanding.
```
If different methods, sources, or theories all point to the same conclusion → confidence increases. If they conflict → something hidden, biased, or manipulated is happening.
```

### Fundamental problem
Single-method thinking = fragile truth
1. Human lie (intentionally or unconsciously)
2. Method distort reality (psychological bias)
3. Researcher have bias
4. Social reality is complex and contextual


### Core Types
#### 1. Data Triangulation
Different data sources, same phenomenon  
Example (emotional dependence):
1. Interview partner A
2. Observe couple interaction
3. Analyze chat messages
4. Third-party report (friend/therapist)

If all suggest dependence → strong evidence  
If only self-report shows it → suspicious  

Used when: studying behavior, relationships, power dynamics

#### 2. Methodological Triangulation
Different methods, same question  
Methods:
1. Interviews (subjective meaning)
2. Surveys (patterns)
3. Observation (actual behavior)
4. Experiments (causality)

Example (manipulation):
1. Survey: "Do you feel pressured?"
2. Observation: partner interrupts, guilt-trips
3. Experiment: withdrawal of affection → anxiety spike

If methods converge → manipulation likely
```
Manipulators often pass surveys but fail behavioral observation.
```

#### 3. Theoretical Triangulation
Different theories explain the same data  
Example: emotional dependence
1. Attachment theory → anxious attachment
2. Power-dependence theory → asymmetrical resources
3. Behavioral conditioning → intermittent reinforcement
4. Psychodynamic → fear of abandonment
   
If multiple theories independently explain the same pattern, the phenomenon is robust.
If only one theory fits → theory bias risk.

#### 4. Investigator Triangulation
Different researchers interpret the same data, Why? :
1. Interpretation is subjective
2. Especially dangerous in qualitative research

Example:
1. Researcher A sees "care"
2. Researcher B sees "control"
3. Researcher C sees "dependency reinforcement"

The overlap is the strongest signal.


### Triangulation Logic
Triangulation is NOT about agreement only. There are 3 outcomes, and each means something.
##### 1. Convergence
All sources align shows high confidence & stable finding
##### 2. Complementarity
Each method reveals a different layer (deeper understanding). Example: 
1. interview: "I love them"
2. Observation: fear-based compliance
3. Theory: trauma bonding
This is common in manipulation research.
##### 3. Contradiction
Methods conflict:
- This is NOT failure
- This is where truth hides  

Example:
1. Self-report: "I'm happy"
2. Behavior: anxiety, avoidance
3. Physiological: stress response
Contradiction = hidden coercion, denial, or unconscious processes

### Advanced triangulation
#### a) Temporal triangulation
Same phenomenon across time (Before / during / after relationship) & Pattern stability matters more than moments

#### b) Level triangulation
1. Individual (psychology)
2. Dyadic (relationship)
3. Structural (social norms, power)
Example:
Emotional dependence isn't just "personality", It's interaction + power + environment.

#### c) Signal vs Noise logic
Advanced researchers ask:
1. Which methods are harder to fake?
2. Which signals are unconscious?
3. Which data costs the subject effort?
   
Behavior > physiology > language

#### Conclusion
```
Triangulation is the systematic use of multiple independent perspectives to detect truth, expose bias, and reveal hidden power dynamics.
```
