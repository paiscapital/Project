# Uses & Gratifications Theory

Why people actively seek out specific media & what needs or gratifications they get from it. People actively choose media to satisfy specific needs or desires.
```
People dont just consume media. They actively choose what they consume to satisfy specific needs, like learning, entertainment, socializing, or escaping stress
```

### Assumptions
1. Active audience : People select media consciously.
2. Goal oriented : People use media to fulfill social & psychological needs.
3. Awareness : People know what they want & why.
4. Competition : Media compete with other sources (friends, hobbies) to satisfy the same needs.
5. Audience control meaning : Media effects depends on how people interprets it.

### UGT Needs
1. Cognitive : Acquire info, knowledge, understanding (course)
2. Affectives : Emotional, aesthetic, experience (movie)
3. Personal integrative : Boosting crediblity, identity (Posting media)
4. Social integrative : Maintain relationship (stay connected)
5. Tension release : Escaping real life stress (gaming)

### Process
1. Identify a need – A person feels curious, bored, lonely, or stressed.
2. Select media – They choose the media that best fulfills that need.
3. Use media – They engage with the content.
4. Gratification obtained – They feel entertained, informed, connected, or relieved.
5. Feedback loop – Satisfaction influences future media choices.

### Relevance
- Social Media: People choose platforms based on what they want (news, entertainment, social connection).  
- Streaming Services: Netflix or YouTube users actively pick content based on mood or interest.
- Gaming: Players select games for challenge, socialization, or stress relief.
  
Modern research extends U&G to mobile apps, social networks, and virtual communities. It helps marketers, media planners, and social scientists understand why people engage with media voluntarily.
