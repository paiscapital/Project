# Moral Panic Theory

MPT refer to a situation in which a group, behavior, idea and phenomenon is potrayed by the media as a threat to societal values, norms, public safety even though the actual threat is often exaggerated to reality. or when society becomes extremely afraid of something that is exaggerated or misunderstood, and that fear spreads faster than facts.

### Formula 

##### Trigger event
```
→ Media exaggeration
→ Public fear
→ Moral outrage
→ Authority reaction
→ Long-lasting damage
```
##### Example
1. One student brings a knife
2. Rumor spreads: “Students are becoming violent”
3. Parents panic
4. Media says: “Schools are unsafe”
5. School bans bags, installs metal detectors
6. Most students were never dangerous
##### More 
1. Claim: “Music corrupts youth”
2. Reality: Normal cultural change
3. Result: Bans, arrests, moral campaigns

#### Core Elements
```
1. Created -> A group or thing becomes the villain (Teenagers, Immigrants, Video games, Music genres, Social media, New technology) They are blamed for “moral decline.”   

2. Media blow up -> Fear sells attention : Selects extreme cases, Repeats emotional language, Removes context using word  (crisis, threat, out of control, destroying our values)  

3. Public fear spreads socially -> Share stories emotionally, Don’t verify facts, Feel “something must be done”  

4. Authorities respond strongly -> Politicians, schools, police react by: Creating new laws, Increasing surveillance, Banning things  

5. The panic fades—but damage remains -> Rights restricted, Groups stigmatized, Policies stay  



#### Psychology Side 
1. Fear of Change : New things feel dangerous 
2. Need for Control : Society wants simple villains (Place to blame)
3. Emotion beats statistics : People feel stories, not data
4. Us vs Them thinking : Blaming an “other” unites the group

#### Moral panic vs real danger
| Real Problem          | Moral Panic      |
| --------------------- | ---------------- |
| Evidence-driven       | Emotion-driven   |
| Proportional response | Overreaction     |
| Long-term solutions   | Quick punishment |
| Data-based            | Story-based      |
