# Attention Economy

The Attention Economy is an economic and communicative system in which human attention functions as a scarce and tradable resource.    
It describes how information producers, digital platforms, and media technologies compete to capture, hold, and monetize the limited cognitive (process of acquiring knowledge and understanding through thought, experience, and the senses) focus of individuals, often through algorithmic personalization and behavioral data optimization.
> Attention = Currency  
> The attention economy is the market where human focus (scarce: Time->1h->media zero sum game) is the product (Ads, Recommended, Track behavior), digital platforms are the brokers (connect buyer & seller), algorithms are the traders decide which content to show to which user, at what time, in order to maximize watch time (Algorithmic culture), and advertisers are the buyers (Companies pay money to access your attention).

#### Example flow with Lasswell Model
| Element | classic | Attention economy |
|---------|---------|-------------|
| Who/Sender | Human Communicator (Journalist, speaker) | Algorithmic Communicator (platform that host monitize & distributes, model based on data pattern) |
| Says What/Message | Crafted message (speech,article,image) | Personalized content optimized for emotional engagement (based on user history) |
| In which channel | TV, Radio, newspaper etc | Platform ecosystem (social media, app, feed) |
| To whom (Audience) | Passive receiver | Data-profiled user segment (group of users clustered by shared behavioral, demographic, psychographic, or contextual data patterns, identified by algorithms) |
| With what effect | Persuasion, knowledge, influence | Captured attention, data generation, behavioral modification |

How behavioral modification?  
> Behavior is shaped by user history and continuously re-shaped by algorithmic feedback. The data does not merely reflect human behavior — it trains it.

Example:
1. You watch a few videos about psychology.  
2. TikTok’s algorithm notices you seem curious and thoughtful.  
3. It starts showing you more emotional or deep videos that match what you liked before.  
4. You watch even longer — TikTok sees that it worked.  
5. So it keeps giving you similar emotional content again and again.  
6. After some time, your whole For You Page is filled with deep or emotional videos.  
7. Slowly, your interests and the way you see yourself start to change because of what you keep watching.

```python
import spacy
import itertools
import lightgbm as lgb
import pandas as pd
import numpy as np
from concurrent.futures import ThreadPoolExecutor

history = {
    'Faiz':[
        "how to use finance monte carlo", "how to calculate montecarlo", "AI video meme", "The best philosopher ever"
    ], # Finance ai philosophy
    'Racheal':[
        "How to cook sandwich", "How to make a spagheti", "How to make strawberry juice"
    ] # cooking
}

class algoCult:

    def __init__(self, history):
        self.history = history
        self.nlp = spacy.load("en_core_web_sm")

    def keyPoint(self, text):
        proc = self.nlp(text)
        token = [token for token in proc if token.pos_ in ("NOUN", "PROPN") and not token.is_stop]
        if not token:
            token = [token for token in proc if token.pos_ in ("ADJ", "VERB") and not token.is_stop]

        if token:
            ranked = sorted(token, key=lambda t: (t.is_title, len(text)), reverse=True)
            return ranked[0].lemma_.lower()

        return None

    def convert_(self, history):
        newHist = {}
        with ThreadPoolExecutor() as execute:
            for key, value in history.items():
                newHist[key] = list(execute.map(self.keyPoint, value))

        return newHist

algo = algoCult(history)
newHist = algo.convert_(history)

# Get algocult list -> optimize/only choose content top 1 -> Monitize/ sell to companies
contents = {
    'finance': [
        {"id": 1, "title":"Finance for beginner", "views": 1000, "likes": 120, "comments": 15, "shares": 10, "age_hours": 2},
        {"id": 2, "title":"How can finance be top industries", "views": 5000, "likes": 300, "comments": 20, "shares": 5, "age_hours": 24},
        {"id": 3, "title":"Finance is hard as f*", "views": 700, "likes": 90, "comments": 5, "shares": 2, "age_hours": 1}
    ],
    'ai': [
        {"id": 1, "title":"AI is make our cult efficient", "views": 19000, "likes": 120, "comments": 15, "shares": 10, "age_hours": 2},
        {"id": 2, "title":"AI is bullshit", "views": 50080, "likes": 300, "comments": 20, "shares": 5, "age_hours": 24},
        {"id": 3, "title":"AI will take over the world as f*", "views": 7000, "likes": 970, "comments": 5, "shares": 2, "age_hours": 1}
    ],
    'philosopher':[
        {"id": 1, "title":"Why socrates is bad", "views": 1000, "likes": 120, "comments": 15, "shares": 10, "age_hours": 2},
        {"id": 2, "title":"How socrates died", "views": 5000, "likes": 300, "comments": 20, "shares": 5, "age_hours": 24},
        {"id": 3, "title":"Diagones is funny", "views": 700, "likes": 90, "comments": 5, "shares": 2, "age_hours": 1}
    ],
    'sandwich': [
        {"id": 1, "title":"Sandwich is delicoius", "views": 1000, "likes": 120, "comments": 15, "shares": 10, "age_hours": 2},
        {"id": 2, "title":"Sandwich uwu", "views": 5000, "likes": 300, "comments": 20, "shares": 5, "age_hours": 24},
        {"id": 3, "title":"F* with sandwich", "views": 700, "likes": 90, "comments": 5, "shares": 2, "age_hours": 1}
    ],
    'spagheti':[
        {"id": 1, "title":"I love soagheti", "views": 1000, "likes": 120, "comments": 15, "shares": 10, "age_hours": 2},
        {"id": 2, "title":"Is spagheti is healthy", "views": 5000, "likes": 300, "comments": 20, "shares": 5, "age_hours": 24},
        {"id": 3, "title":"Is spaghet*", "views": 700, "likes": 90, "comments": 5, "shares": 2, "age_hours": 1}
    ]
}


def key_at_index(d, index):
    return next(itertools.islice(d, index, index+1))

def optimizeProfile(contents, history):
    df = pd.concat([
        pd.json_normalize(items).assign(topic=topic)
        for topic, items in contents.items()
    ])

    df["time_decay"] = np.exp(-df["age_hours"] / 12) # latest post testing boost, old one expired
    df["engagement"] = (
        0.4 * df["views"] +
        0.3 * df["likes"] +
        0.2 * df["comments"] +
        0.1 * df["shares"]
    ) / (df["age_hours"] + 1)

    features = ["views", "likes", "comments", "shares", "time_decay"]
    df[features] = (df[features] - df[features].min()) / (df[features].max() - df[features].min())
    group_sizes = df.groupby('topic').size().tolist() # [3, 3, 3, 3, 3]

    train_ = lgb.Dataset(df[features], label=pd.qcut(df["engagement"], q=5, labels=False, duplicates="drop"), group=group_sizes)

    # lgb params needed
    params = {
        "objective": "lambdarank",   # learning-to-rank
        "metric": "ndcg",
        "boosting_type": "gbdt",
        "num_leaves": 15,
        "learning_rate": 0.1,
        "min_data_in_leaf": 1,
        "verbose": -1
    }
    
    model = lgb.train(params, train_, num_boost_round=50)
    df["attention_score"] = model.predict(df[features])
    ranked = df.sort_values("attention_score", ascending=False).reset_index(drop=True)

    profile = {user: [] for user in history.keys()}
    for _, row in ranked.iterrows():
        for user, topics in history.items():
            if row['topic'] in topics:
                profile[user].append({
                    'topic': row['topic'],
                    'data': {
                        'id': row['id'],
                        'title': row['title'],
                        'attention_score': float(row['attention_score'])
                    }
                })
            
    return profile
    
optimizeProfile(contents, newHist)

```
