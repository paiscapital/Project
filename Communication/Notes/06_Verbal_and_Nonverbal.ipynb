# Verbal & Nonverbal Communication

## Verbal (data)
Use word or linguistics symbols includes spoken, written.

#### 1) Layers
##### (a) Lexical layer (words)
The lexical layer refers to the choice of words (vocabulary) a speaker or writer uses. Words carry meaning, emotion, social values, and power. Different word choice can change tone, formality, politeness, authority or emotional impact.   


Vocabulary choice is unconscious selection of words from all possible alternatives (context, audience, purpose, emotional sensitivity, culture and norms).  

1. Technical words:
   ```
   Specialized terms used in professional, academic, medical, legal, or scientific contexts
   ```  
   
2. Everyday words:
   ```
   Everyday common, informal, socially softened words used in daily communication
   ```

##### (b) Syntatic layer (structure)
Syntatic layer focusses on how words are arranged into sentences. Syntax shape meaning, power, responsiblity, and emotion.   

  
###### 1. Word order
-  Agency (whois doing the action)
-  Focus (what the sentence wants you to notice)
-  Responsiblity (who is blamed or protected)


###### 2. Active structure
- Active voice: "You broke the rule"
- Structure: Subject=You, Verb=broke, Object=The rule
- Effect: Direct, Assigns blame clearly, personal accountability


###### 3. Passive structure (Blame removal/Normalization)
- Passive voice: "The rule was broken"
- Structure: Subject=The rule, Verb=was broken, Agent=removed or optional
- Effect: Blame is errased, responsibility become vagues, sounds formal/neutral/institutional.

##### (c) Semantic layer (meaning)
What a word, phrase, sentence actually means. The relationship between language and meaning.    
Meaning is not fixed inside words, instead meaning emerges from:
1. The word used
2. How they are combined
3. The concept they refer to
4. The context assumed by the listener

##### Semantic Ambiguity
A word, phrase, sentence allows more than one possible meaning. This is a semantic problem, not a grammar mistake because:
1. Word have multiple meanings (polysemy)
2. Different grammatical roles share the same form
3. Missing contextual information 


###### a) Lexical Ambiguity
One word with multiple meaning.  
Example: 
- Duck (animal/action) : I saw her duck

###### b) Structural ambiguity
Meaning depends on how words are grouped.  
Example:
"Old men and women”
- Old men + women
- Old men + old women

###### c) Referential ambiguity
Unclear reference.  
Example: "Ali told Ahmad that he was late"  
- Who is "he"?

#### (d) Pragmatic layer (intention & context)
What the speaker intends to do by saying it, in a specific situation. Meaning in use.  

The same sentence can mean opposite things. Most conflicts happen even when grammar and vocabulary are correct. Humans don't speak just to describe reality, we speak to act.  

##### 1) Intention (why is it said?) 
What action is the speaker performing?  
Possible intentions:
1. Showing interest
2. Expressing doubt
3. Soft rejection
4. Sarcasm
5. Power signaling
6. Politeness strategy

Language is an action, not just information.

##### 2) Speaker identity & power (Who says it?)
Meaning changes depending on:
1. Authority (boss vs friend)
2. Relationship (lover vs stranger)
3. Expertise (expert vs novice)
4. Culture & personality

Same words, different speaker → different force.

##### 3) Timing (What it is said?)
Timing reveals intention:
1. Immediate response vs delayed response
2. Said before decision vs after decision
3. Said publicly vs privately

Late or vague responses often signal avoidance or rejection.

##### 4) Setting & audience (Where & to whom?)
Context shapes meaning:
1. Formal vs informal
2. Public vs private
3. One-on-one vs group
4. Cultural norms

People adjust speech to save face and manage impressions (people soften words, avoid direct refusal, use ambiguity, perform politenes).

### 2) Why verbal?
Informative (transfering facts, teaching), Expressive (Sharing emotions, attitudes), Persuasive (Changing beliefs, behavior), Phatic/social bonding (Maintaining relationship), Performative (speech act theory). 

#### Power
Language is not neutral.  

##### (a) Control through framing
- "Illegal immigrant" vs "Undocumented worker"
- "Collateral damage" vs "Civilian deaths"
Same reality, different moral perception.

##### (b) Institutional language
1. Legal language
2. Bureaucratic language
3. Academic jargon
The purpose is precision, authority & Exclusion of outsiders.

##### (c) Symbolic dominance (Bourdieu)
Those who control proper language & educated speech, Often control legitimacy, credibility & social hierarchy.

#### Advanced Level
##### (i) Language as symbolic
Words are symbols, not reality itself.  
Example:  
"Death" ≠ the physical event, It's a symbol representing that event. So "He died", "He passed away", "He expired".  
Same reality → different symbols → different emotional effects. This is why propaganda, politics, media framing work.

#### (ii) Language has rules
Language operates on multiple rule systems: Phonological (sound), Syntatic (structure), Semantic (meaning), Pragmatic (context).
  
#### (iii) Language is subjective
Even "objective" words are filtered through:
1. Culture : Direct & Indirect word
2. Experience : Doctor=workplace/People=neutral
3. Emotion : High tone when say "yes" is "angry"/"excited"
4. Ideology : Capitalism=freedom/exploitation depends

That's why misunderstandings happen even when grammar is correct.
  
  
Humans rarely communicate directly, because we care about:
1. Politeness
2. Hierarchy
3. Emotional impact
4. Social consequences

This leads to:
1. Indirect speech acts
2. Implicatures (implied meaning)
3. Euphemisms
4. Strategic ambiguity

Example:
1. "We'll see" → often means "No"
2. "Up to you" → often means "Choose what I want"

## Nonverbal (metadata)
Conveying meaning without words or use nonlinguistics (Physical, aesthetic, signs, symbolic). If verbal and nonverbal cues conflict, nonverbal usually wins. People rely more on nonverbal cues than words when interpreting intention.  

### 1. Types
#### a) Kinesics (Body movement)
It includes posture, gestures, head movements & walking style. Examples:
1. Open posture → confidence, approachability
2. Closed posture → defensiveness, anxiety
3. Mirroring → rapport building (often unconscious)
 
Meaning depends on context, not isolated gestures.  

#### b) Facial expression (Emotional signals)
Universally recognized emotions such as happiness, sadness, anger, fear, surprise & disgust.
Micro-expressions:
1. Last fractions of a second
2. Occur before conscious facial control
3. Reveal genuine, hidden emotions
4. Used in interrogation and psychology to detect true emotional reactions
  
Example:
A smile without eye muscle movement = social or fake smile.

#### c) Eye behavior (Oculesics)
It includes eye contact, gaze direction, blink rate & pupil dilation

Examples:
1. Prolonged eye contact → dominance or intimacy
2. Avoiding eye contact → submission, discomfort, or cultural norm
3. Faster blinking → stress or cognitive load

Culture matters a lot here.  

#### d) Paralanguage (voice without words)
Not what is said, but how the tone, pitch, volume, speed & pauses.  
Example:
Same sentence, different tone → different meaning
1. "Fine." (calm)
2. "Fine." (angry)

Silence it self is communication.  

#### e) Proxemics (use of space)
Personal distance zones:
1. Intimate
2. Personal
3. Social
4. Public

Violating space can signal dominance, aggression, or closeness.  
Example:
A boss standing too close vs a stranger standing too close → very different meanings.

#### f) Haptics (touch)
Types:
1. Handshake
2. Pat on the back
3. Hug

Meaning depends on:
1. Relationship
2. Culture
3.Situation

Touch can:
1. comfort
2. control
3. intimidate

#### g) Chronemics (use of time)
Time communicates value:
1. Being late → power or disrespect
2. Making others wait → dominance
3. Quick response → urgency or care

In institutions, time = authority.

#### h) Appearance & artifacts
It includes clothing, grooming, accessories & objects you use.  
Example:
1. Uniform → authority
2. Minimalist style → control, discipline
3. Luxury items → status signaling

This is symbolic nonverbal communication.


#### 2) Why nonverbal
Nonverbal cues can be used to build trust, assert dominance, create intimidation & signal alliance.  
Examples:
1. Mirroring to create comfort
2. Strategic silence to pressure
3. Controlled stillness to appear powerful

Nonverbal cues are not universal in meaning:
1. Eye contact respectful in West, rude in some Asian cultures
2. Smiling can mean friendliness or embarrassment
3. Silence can mean agreement or resistance

Always interpret through cultural lens.
