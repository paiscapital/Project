# Post Truth Culture

Post-truth culture is a social and cultural condition in which objective facts have less influence on shaping public opinion than appeals to emotion, personal belief, and group identity, resulting in a reality where truth becomes relative, fragmented, and negotiable within different communities.
> People care more about what feels true than what is true.

#### Why It Happen

1. Information Overload:
    - Too much information everywhere (news,posts,memes,videos,opinions) no one can check it all& hard to know what’s real.
    - This make our brains say "This feels right, i will believe this"
    - Feel tired of checking facts& Start trusting what fits their gut feeling.

2. Emotional Politics:
    - Modern media and politics don’t try to convince people with logic anymore.
    - Loss of trust → people doubt governments, media, scientists.
    - They use emotion (fear, anger, pride, or identity) because emotion spreads faster and grabs attention.
    - "Theyre stealing your freedom!" feels more powerful than "Policy X will reduce GDP by  2.4%"

| Mindset | Meaning | Example |
|---------|---------|---------|
| Is it true (Epistemic) | You check the facts& sources | Reading scientific reports |
| Does it feel true (Affective) | You believe what fits your emotions or worldview. | Sharing a meme that “feels right.” |




#### Psychology Side
1. Emotional over Rational Processing:
    - When people see information that fits their beliefs or values, it triggers emotional satisfaction, a feeling of safety or belonging.
    - The brain releases dopamine when something “feels right,” so emotions become a shortcut for deciding what’s true (Resonance)

2. Cognitive Biases:
    - Confirmation bias: We seek information that supports what we already believe.
    - Availability bias: We believe what’s easiest to remember or most repeated.
    - Affective reasoning: We judge ideas based on how they make us feel, not on evidence.

3. Social Identity& Belonging:
    - When a belief becomes part of group identity, disagreeing feels like betrayal.
    - People defend ideas that match their tribe’s emotions, not necessarily the truth.
