# Osgood-Schramm Model

Communication is not one-way, its a continuous loop where both parties encode, decode, and interpret messages simultaneously, circular model of communication.  

  
![OSGOOD](https://www.researchgate.net/profile/Magnus-Bostroem/publication/344016281/figure/fig8/AS:930920727064577@1598960205036/Osgood-Schramm-model-with-overlapping-fields-of-experience.png)

#### Key Points
1. Circular, not linear: Theres no strict sender → receiver. Both parties take turns sending and receiving.
2. Encoding & decoding: Each person interprets messages based on their experiences, context, and understanding.
3. Feedback is constant: You dont just respond after the message ends, your reactions are ongoing.
4. Meaning-centered: Focuses on how people understand messages, not just the message itself.

```
[Person A] ↔ [Message/Interpretation] ↔ [Person B]
```
Each person is both sender and receiver, Communication is continuous and dynamic.

#### Components
##### 1. Encoder / Decoder
Encoder = When you send a message (words, tone, emojis)  
Decoder = When you receive a message (interpret meaning/translator)  

Example: You say 'Hai' to zul, zul receive it and decode it as 'faiz greetings to me'
  
##### 2. Message
The message is not just words, but includes tone, emotion, gestures, context & meaning is constructed, not just transmitted.

##### 3. Feedback
After decoding, each participant reacts based on their interpretation. Feedback is instantaneous and continuous, not just a final response.

##### 4. Shared Meaning
Communication is effective only if theres some overlap in experience or knowledge. Misunderstandings happen when shared experience is low.


### Chat Loop Simulator


````python
import random

class Agent:
    def __init__(self, name, mood):
        self.mood = mood
        self.name =  name

    def encoder(self, message):
        
        if self.mood == "happy":
            return message + ":)"
        elif self.mood == "angry":
            return message.upper() + "!!! >:("
        else:
            return message

    def decoder(self, message):

        if '!' in message:
            return 'Feels angry'
        elif ':)' in message:
            return 'Feels happy'
        else:
            return 'Neutral'

def chatloop(agent1, agent2, messages):
    for msg in messages:
        sent_msg = agent1.encoder(msg)
        print(f"{agent1.name} says: {sent_msg}")
        interpretation = agent2.decoder(sent_msg)
        print(f"{agent2.name} interprets: {interpretation}\n")

        # Swap Role
        agent1, agent2 = agent2, agent1

zul = Agent("Zul", "happy")
pais = Agent("Pais", "angry")
messages = ["Can you finish the report today?", "I think we need a break", "Let's meet tomorrow"]

chatloop(zul, pais, messages)

    
````
