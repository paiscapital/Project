# Dance Helical Model  

DHM represents communication as continuous, cumulative, and evolving process. It visualizes communication as a helix (spiral) that starts at a point and expands outward over time, showing that each interaction builds on past experiences, grows in complexity, and never exactly repeats.

## Core Idea
- Communication is not static, it grows like a spiral expanding outward
- Every interaction builds on previous experience
- communication is continuous, cumulative, and never exactly repeats  

```python
theta = np.linspace(0, 8 * np.pi, 1000)  # angle from 0 to 8
a = 0.1  # starting radius
b = 0.2  # growth rate

r = a + b * theta  # radius increases with angle

# Convert polar to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot
plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, r, color='purple', linewidth=1)

ax.set_xticks([])
ax.set_yticks([])
ax.grid(False)
ax.spines['polar'].set_visible(False)

ax.text(theta[0], r[0], "Communication", color='black', fontsize=11, ha='center', va='center')
ax.text(theta[2], r[-2], "", color='black', fontsize=20, ha='center', va='center')

# Title
ax.set_title("DHM Spiral", fontsize=16)

plt.show()
```

# Structure
1. Communication as a spiral that starting at a point and widening as it progress.  
2. Communication starts simple with minimal experience.  
3. As it moves forward:
    > People add experience and knowledge
    > Message are shaped by prior interactions
    > Communication expands in scope and depth like a widening spiral
